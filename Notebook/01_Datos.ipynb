```python
# Vemos si todos los id son unicos
print(df1['id'].is_unique)
print(df2['id'].is_unique)

# Vemos si coinciden en los dos dataframes
coinciden = set(df1['id']).intersection(df2['id'])
print(len(coinciden))

# vemos los id que estan solo en un dataframe
solo_df1 = set(df1['id']) - set(df2['id'])
solo_df2 = set(df2['id']) - set(df1['id'])

print("IDs solo en df1:", len(solo_df1))
print("IDs solo en df2:", len(solo_df2))
```

```python
#Cambiamos el nombre del mes es español a numero para poder covertirlo en formato fecha

df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('enero', '01')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('febrero', '02')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('marzo', '03')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('abril', '04')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('mayo', '05')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('junio', '06')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('julio', '07')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('agosto', '08')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('septiembre', '09')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('octubre', '10')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('noviembre', '11')
df['fecha_interaccion'] = df['fecha_interaccion'].str.replace('diciembre', '12')
```

```python
# Creamos una columna con el tiempo que lleva siendo cliente hasta la interaccion en meses

df['tiempo_cliente'] = (df['fecha_interaccion'] - df['fecha_alta_cliente']).dt.days // 30
```

```python
df['tiempo_cliente'].value_counts()
```

```python
df['tiempo_cliente'].describe()
```

```python
# los segmentamos por nuevo, reciente y antiguo

def segmentacion_clientes(meses):
    if meses < 6 :
        return 'nuevo'
    elif meses < 24:
        return 'reciente'
    else:
        return 'antiguo'
    
df['antiguedad_cliente'] = df['tiempo_cliente'].apply(segmentacion_clientes)

```

```python
df['antiguedad_cliente'].value_counts()
```

```python
# Creamos la columna mes de interaccion

df['mes_interaccion'] = df['fecha_interaccion'].dt.month
```

```python
df['mes_interaccion'].value_counts()
```

```python
mes_año = {1 : 'enero',
           2 : 'febrero',
           3 : 'marzo',
           4 : 'abril',
           5 : 'mayo',
           6 : 'junio',
           7 : 'julio',
           8 : 'agosto',
           9 : 'septiembre',
           10 : 'octubre',
           11 : 'noviembre',
           12 : 'diciembre'}
```

```python
df['mes_interaccion'] = df['mes_interaccion'].map(mes_año)
```

```python
df['mes_interaccion'].value_counts()
```

```python
# y el dia de la interaccion

df['dia_interaccion'] = df['fecha_interaccion'].dt.day_of_week
```

```python
df['dia_interaccion'].value_counts()
```

```python
dias_semana = {0 : 'lunes',
               1 : 'martes',
               2 : 'miercoles',
               3 : 'jueves',
               4 : 'viernes',
               5 : 'sabado',
               6 : 'domingo'}
```

```python
df['dia_interaccion'] = df['dia_interaccion'].map(dias_semana)
```
